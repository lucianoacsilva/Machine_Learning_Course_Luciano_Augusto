# Thomson Sampling

Notebook with a template implementation of Thomson Sampling Algorithm for reinforcement learning

## Importing Dataset

```r
dataset = read.csv('Ads_CTR_Optimisation.csv')
```

## Implementation of the algorithm

### Setting initial variables

```r
N = 10000
d = 10

numberRewards0I = integer(d)
numberRewards1I = integer(d)
totalReward = 0
adsSelected = integer(0)

```

### Implementation itself

```r
for (n in 1:N) {
  ad = 0
  maxRandom = 0
    
  for (i in 1:d) {
    randomBeta = rbeta(n = 1, shape1 = numberRewards1I[i] + 1, shape2 = numberRewards0I[i] + 1)
    if (randomBeta > maxRandom) {
      maxRandom = randomBeta
      ad = i
    }
  } 

  adsSelected = append(adsSelected, ad)
  reward = dataset[n, ad]
    
  if (reward ==1) {
      numberRewards1I[ad] = numberRewards1I[ad] + 1 
  }
    
  else {
      numberRewards0I[ad] = numberRewards0I[ad] + 1
  }
    
  totalReward = totalReward + reward
}
```

### Visualizing the results

```r
hist(adsSelected,
     col = 'purple',
     main = 'Histogram of Selections',
     xlab = 'Advertisements',
     ylab = 'Number of Selections')
```
